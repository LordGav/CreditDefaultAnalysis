```python
y = df['default.payment.next.month']
y.head()
```

```python
# one hot encoding for 'SEX', 'MARRIAGE' & 'EDUCATION'
X = pd.get_dummies(data=df, columns=['SEX', 'MARRIAGE', 'EDUCATION'])
X.drop(columns=['default.payment.next.month'], axis=1, inplace=True)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import xgboost as xgb

model = xgb.XGBClassifier(objective='reg:squarederror', learning_rate = 0.1, max_depth = 5, n_estimators = 100,
                          booster="gbtree")
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)
print(f"Accuracy: {100*accuracy_score(y_pred, y_test):.2f}%")
```

```python
pd.DataFrame(confusion_matrix(y_pred, y_test))
```

```python
print(classification_report(y_test, y_pred))
```
